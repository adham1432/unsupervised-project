```python
import numpy as np
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/winequality-white.csv') # unlabel data 4898 rows Ã— 12 columns
print(data)
```

```python
mean = data.mean(axis=0)  # Calculate mean of each columns
print("mean:")
print(mean)
```

```python
data_shifted = data - mean
print(data_shifted)
```

```python
cov_matrix=np.cov(data_shifted, rowvar=False)
print("cov_matrix:")
print(cov_matrix)
print("cov_matrix.shape")
print(cov_matrix.shape)
```

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_matrix)
print ("eigen_values:")
print (eigen_values)
print("eigen_values.shape")
print(eigen_values.shape)
print ("-----------------------------------------------------------------------")
print ("eigen_vectors:")
print (eigen_vectors)
print("eigen_vectors.shape")
print(eigen_vectors.shape)
```

```python
sorted_indices = np.argsort(eigen_values)[::-1]
eigenvalues = eigen_values[sorted_indices]
eigenvectors = eigen_vectors[:, sorted_indices]
print ("eigen_values:")
print (eigenvalues)
print("eigen_values.shape")
print(eigenvalues.shape)
print ("-----------------------------------------------------------------------")
print ("eigen_vectors:")
print (eigenvectors)
print("eigen_vectors.shape")
print(eigenvectors.shape)
```

```python
import numpy as np

# Assuming `data_shifted` is your mean-centered data
# Assuming `eigenvectors` are the principal components

explained_variances = []
total_variance = np.sum(np.var(data_shifted, axis=0))  # Total variance of the original data

for k in range(1, data.shape[1] + 1):  # Iterate through 1 to the number of columns
    Q = eigenvectors[:, :k]
    print(f"Q for k={k}:\n{Q}\n")  # Print Q for demonstration

    # Project data onto the first k eigenvectors
    data_reduced = data_shifted @ Q
    print(f"data_reduced for k={k}:\n{data_reduced}\n")  # Print data_reduced for demonstration

    # Compute variance explained by these k components
    reduced_variance = np.sum(np.var(data_reduced, axis=0))  # Sum of variances of reduced data
    explained_variance_ratio = reduced_variance / total_variance
    explained_variances.append(explained_variance_ratio)
    print(f"Explained variance for k={k}: {explained_variance_ratio:.2%}\n")

# Find the smallest k where explained variance >= 95%
threshold = 0.95
best_k = np.argmax(np.array(explained_variances) >= threshold) + 1
print(f"The best number of components (Q) is: {best_k}")
```

```python
# Choose the top n_components eigenvectors for dimensionality reduction
n_components = 2 # Or however many components you want to keep
Q = eigenvectors[:, :n_components]

# Then, calculate X_reduced:
data_reduced =data_shifted @ Q  # Use matrix multiplication with the correct order


print(data_reduced)
```

```python
data=np.array(data)
data_reduced=np.array(data_reduced)
```

```python
#inlization membership
num_cluster=4
g=1.25 # Fuzziness parameter

ms=np.random.rand(data.shape[0],num_cluster) # Randomly initialize membership matrix
ms/=np.sum(ms, axis=1)[:,np.newaxis]
ms
```

```python
def calculate_centroid (data, num_cluster, ms, g):
    Centroids=np.zeros((num_cluster,data.shape[1]))
    for i in range (num_cluster):
        Centroids[i,:]=np.sum((ms[:,i]**g)[:,np.newaxis]*data, axis=0)/np.sum(ms[:,i]**g)
    return Centroids
```

```python
def update_membership (data, Centroids, num_cluster , g):
    ms_new=np.zeros((data.shape[0],num_cluster))
    for i in range (num_cluster):
        ms_new[:,i]=np.linalg.norm(data-Centroids[i,:],axis=1) #Calculate distance of each data point to the current centroid

    ms_new=1/ (ms_new ** (2/(g-1)) * np.sum((1/ms_new) ** (2/(g-1)) , axis=1 )[:,np.newaxis] ) # Update membership values using the fuzzy clustering formula
    return ms_new
```

```python
#cluster orignal data (before PCA)
max_iteration=100
for iteration in range (max_iteration):
    Centroids=calculate_centroid(data, 4, ms , 1.25)
    ms_new= update_membership (data,Centroids, 4 , 1.25 )



    if np.linalg.norm (ms_new- ms)<=0.00001:
        break
    ms=ms_new

    labels=np.argmax(ms_new, axis=1)
```

```python
#cluster data redued (after PCA)
max_iteration=100
for iteration in range (max_iteration):
    Centroids=calculate_centroid(data_reduced, 4, ms ,1.25 )
    ms_new= update_membership (data_reduced,Centroids, 4 , 1.25 )



    if np.linalg.norm (ms_new- ms)<=0.00001:
        break
    ms=ms_new

    labels2=np.argmax(ms_new, axis=1)# sample is bigest membership
```
